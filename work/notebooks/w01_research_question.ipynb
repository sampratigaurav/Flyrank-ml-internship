## 1. My lane (or freestyle) and why

**Chosen Lane:** Refresh / Content Opportunity Scoring

**Why this lane?**
In SEO and content management, not all pages are created equal. Some pages draw massive traffic but are slowly decaying because they haven't been updated recently. Re-writing or updating a page ("refreshing") takes significant editor time and resources. By building an opportunity scoring model, we can prioritize pages that are high-value (visible) but decaying (stale), maximizing the return on content investment.

## 2. The question: decision, action, cost of a wrong call

- **What decision does this improve?**
  It improves the content team's decision of *which existing content piece to refresh or rewrite first*.
- **Who acts on the output, and what do they do?**
  The Content Editor/Manager acts on the output. They will take the top-ranked pages from our queue and assign them to writers for updates, content expansion, or keyword optimization.
- **What does a wrong recommendation cost?**
  A false positive (recommending a page that is actually stable and doesn't need a refresh) costs wasted editor hours and writer fees (approximately 3–5 hours per article). A false negative (missing a page that is declining rapidly) costs lost search impressions, resulting in a drop in organic traffic and lead conversions for the client.
- **Why does data/ML help at all?**
  A simple static rule (e.g., "refresh anything older than 180 days") is too crude because it doesn't account for the complex interactions between content age, historical search impressions, keyword rank decay, and GA4 engagement signals. ML can weight these factors dynamically to generate a highly prioritized queue.

## 3. Quick look at the data (2-3 real numbers)

Let's load the starter CSV dataset and calculate a few baseline numbers to support our lane choice.

```python
import pandas as pd
import numpy as np

# Load the starter CSV dataset
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

# 1. Base decline rate in the dataset
base_decline_rate = (df["trend_direction"].str.lower() == "down").mean()
print(f"1. Overall content decline rate (base rate): {base_decline_rate * 100:.2f}%")

# 2. Staleness vs Decline
avg_days_decline = df[df["trend_direction"].str.lower() == "down"]["days_since_last_update"].mean()
avg_days_stable = df[df["trend_direction"].str.lower() == "stable"]["days_since_last_update"].mean()
print(f"2. Mean days since update for declining pages: {avg_days_decline:.2f} days")
print(f"   Mean days since update for stable pages: {avg_days_stable:.2f} days")

# 3. High Impressions but Stale
stale_visible = df[(df["days_since_last_update"] >= 90) & (df["impressions_90d"] >= 500)]
decline_rate_stale_visible = stale_visible["trend_direction"].str.lower().eq("down").mean()
print(f"3. Number of high-traffic (>=500 impressions), stale (>=90 days) pages: {stale_visible.shape[0]}")
print(f"   Decline rate among this group: {decline_rate_stale_visible * 100:.2f}%")
```

## 4. Careful words: what I can and can't claim

- **What we CAN claim:**
  We can claim that our model identifies *decision-support* recommendations and *directional* signals. We can show that pages matching certain patterns (e.g., high staleness combined with high search impressions) are statistically *observed* to have a higher rate of decline, which helps prioritize editor attention.
- **What we CANNOT claim:**
  We cannot claim any *causal proof* (e.g., "Updating this page will cause a 20% traffic increase") or that we are "predicting Google's ranking algorithm." Google's ranking factors are proprietary and constantly shifting; our model simply predicts performance outcomes on the client's own history.
